```python
data = ''
with open("/DS/DSG/work/newlogs/without_updates/1-64.txt") as f:
    data = f.read()
```

```python
import re

iterations = data.split('END\n')[:-1]
metrics = []
regex_loading_time = r"Graph loading time \(in ms\): (\d+)"
regex_writer_thread = r"Number of writer Threads: (\d+)"

for iteration in iterations:
    m = re.search(regex_loading_time, iteration)
    # print(m[0], m[1])
    loading_time = int(m[1])
    m = re.search(regex_writer_thread, iteration)
    # print(m[0], m[1])
    writer_threads = int(m[1])
    print(writer_threads, loading_time)
```
